```python
import pandas as pd
import nltk
```

```python
Regis = pd.read_pickle('St. Regis.pkl')
JW = pd.read_pickle('JW.pkl')
TajLands = pd.read_pickle('TajLands.pkl')
TajMahal = pd.read_pickle('TajMahalTower.pkl')
Leela = pd.read_pickle('TheLeela.pkl')
```

# Analysing all the reviews 

# Cleaning the data

```python
Regis_review =  Regis.Reviews
JW_review = JW.Reviews
Lands_review = TajLands.Reviews
Mahal_review = TajMahal.Reviews
Leela_review = Leela.Reviews
```

### Word Tokenization

```python
Regis_review =  [' '.join(nltk.word_tokenize(i)) for i in Regis_review]
JW_review =  [' '.join(nltk.word_tokenize(i)) for i in JW_review]
Lands_review =  [' '.join(nltk.word_tokenize(i)) for i in Lands_review]
Mahal_review =  [' '.join(nltk.word_tokenize(i)) for i in Mahal_review]
Leela_review =  [' '.join(nltk.word_tokenize(i)) for i in Leela_review]
```

### Case Conversion

```python
Regis_review = [i.lower() for i in Regis_review]
JW_review = [i.lower() for i in JW_review]
Lands_review = [i.lower() for i in Lands_review]
Mahal_review = [i.lower() for i in Mahal_review]
Leela_review = [i.lower() for i in Leela_review]
```

### Removing Special Characters

```python
import re
import string
```

```python
def remove_characters_after_tokenization(tokens):
    pattern = re.compile('[{}]'.format(re.escape(string.punctuation)))
    filtered_tokens = ''.join(filter(None, [pattern.sub('', token) for token in tokens]))
    return filtered_tokens
```

```python
Regis_review = [remove_characters_after_tokenization(token) for token in Regis_review]
JW_review = [remove_characters_after_tokenization(token) for token in JW_review]
Lands_review = [remove_characters_after_tokenization(token) for token in Lands_review]
Mahal_review = [remove_characters_after_tokenization(token) for token in Mahal_review]
Leela_review = [remove_characters_after_tokenization(token) for token in Leela_review]
```

### Expanding Contractions

```python
from contractions import contractions_dict
```

```python
c_re = re.compile('(%s)' % '|'.join(contractions_dict.keys()))
def expandContractions(text, c_re=c_re):
    def replace(match):
        return contractions_dict[match.group(0)]
    return c_re.sub(replace, text)
```

```python
Regis_review = [expandContractions(i) for i in Regis_review]
JW_review = [expandContractions(i) for i in JW_review]
Lands_review = [expandContractions(i) for i in Lands_review]
Mahal_review = [expandContractions(i) for i in Mahal_review]
Leela_review = [expandContractions(i) for i in Leela_review]
```
